```python
def outlier_detection(x):
    q1=df[x].quantile(0.25)
    q3=df[x].quantile(0.75)
    iqr=q3-q1
    cnt=df[(df[x]> (q3+1.5*iqr)) | (df[x]<(q1-1.5*iqr))][x].count()
    return cnt
```

```python
for i in num_cols:
    print('count of outliers w.r.t ' +i+ ' is '+ str(outlier_detection(i)))
```

```python
def outlier_rows(x):
    q1=df[x].quantile(0.25)
    q3=df[x].quantile(0.75)
    iqr=q3-q1
    df_outlier=df[(df[x]> (q3+1.5*iqr)) | (df[x]<(q1-1.5*iqr))][[x,'y']]
    return df_outlier
```

```python
# function to plot catgorical columns distribution w.r.t target label
def perc_plot(cat_column):
    plt.figure(figsize=(12,9))
    denom=41176
    ax=sbn.countplot(x=cat_column, hue='y', data=df)
    for bar in ax.patches:
        ax.annotate('{:.1f}%'.format(100*bar.get_height()/denom), (bar.get_x()+0.1, bar.get_height()+5))
    ax.set_yticklabels(map('{:.1f}%'.format, 100*ax.yaxis.get_majorticklocs()/denom))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
    plt.show()
```
